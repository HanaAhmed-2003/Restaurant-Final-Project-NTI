```python

# Function to DETECT and return outliers
def detect_outliers_iqr(df, column):
    # Ensure we are only calculating on numbers
    if df[column].dtype in ['int64', 'float64']:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # The "|" symbol means "OR". We look for values below the lower bound OR above the upper bound
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        return outliers
        
    return pd.DataFrame() # Return empty if not a numeric column

# ====================================================
# 1. Detect Customer Outliers (Age)
# ====================================================

customers['age'] = pd.to_numeric(customers['age'], errors='coerce')

customer_outliers = detect_outliers_iqr(customers.dropna(subset=['age']), 'age')

print(f"--- Customer Age Outliers: {len(customer_outliers)} found ---")
# Print the first few outliers to see what is wrong with them
if len(customer_outliers) > 0:
    print(customer_outliers[['customer_id', 'first_name', 'age']].head())


# ====================================================
# 2. Detect Order Outliers (Total Sales)
# ====================================================

orders['total_sales'] = pd.to_numeric(orders['total_sales'], errors='coerce')

order_sales_outliers = detect_outliers_iqr(orders.dropna(subset=['total_sales']), 'total_sales')

print(f"\n--- Order Total Sales Outliers: {len(order_sales_outliers)} found ---")
# Print the first few outliers
if len(order_sales_outliers) > 0:
    print(order_sales_outliers[['order_id', 'total_sales']].head())
    
# You can also save your detected outliers to a CSV to review them manually!
# order_sales_outliers.to_csv('sales_outliers_report.csv', index=False)
```
